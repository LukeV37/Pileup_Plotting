```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
plt.rcParams.update({'font.size': 16})
```

```python
mu="200"
```

```python
file_names = ["Baseline_MSE_mu"+mu+".pkl","AttnJVT_MSE_mu"+mu+".pkl","MSE_model_mu"+mu+".pkl"]

results = []

for file in file_names:
    with open(file, 'rb') as f:
        results.append(pickle.load(f))
```

```python
x1 = results[0][0]
y1 = results[0][1]

x2 = results[1][0]
y2 = results[1][1]

x3 = results[2][0]
y3 = results[2][1]
```

```python
plt.title("Regression Task Comparison <\u03BC>="+mu)
plt.plot(x1,y1,color='r',label='BaselineNN',linewidth=2)
plt.plot(x2,y2,color='g',label='AttnJVT',linewidth=2)
plt.plot(x3,y3,color='b',label='AttnJVT wTracks',linewidth=2)

plt.ylabel('Fake Rate',loc='top')
plt.xlabel('Signal Efficiency',loc='right')
plt.yscale('log')
plt.legend(loc='upper left')
plt.xlim([0.8, 1])
plt.ylim([5e-5,1])
plt.grid(which='both')
plt.show()
```
