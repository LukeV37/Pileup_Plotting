```python
import uproot
import awkward as ak
import matplotlib.pyplot as plt
```

```python
path_mu60  = "/home/lvaughan/Pileup/Pileup_Project/pythia/output/dataset_diHiggs_mu60_NumEvents20k_MinJetpT25.root:fastjet"
path_mu200 = "/home/lvaughan/Pileup/Pileup_Project/pythia/output/dataset_diHiggs_mu200_NumEvents20k_MinJetpT80.root:fastjet"
```

```python
with uproot.open(path_mu60) as f:
    print(f.keys())
    Efrac_mu60 = f['jet_true_Efrac'].array()
    Mfrac_mu60 = f['jet_true_Mfrac'].array()
```

```python
with uproot.open(path_mu200) as f:
    print(f.keys())
    Efrac_mu200 = f['jet_true_Efrac'].array()
    Mfrac_mu200 = f['jet_true_Mfrac'].array()
```

```python
bins=50
Range=(0,1)
```

```python
plt.title("Efrac Distribution of Jets")
plt.hist(ak.ravel(Efrac_mu60),histtype='step',label='mu60',color='b',range=Range,bins=bins)
plt.hist(ak.ravel(Efrac_mu200),linestyle='--',histtype='step',label='mu200',color='b',range=Range,bins=bins)
plt.yscale('log')
plt.grid(which='both')
plt.ylabel("Counts",loc='top')
plt.xlabel("Efrac", loc='right')
plt.legend()
plt.show()
```

```python
plt.title("Mfrac Distribution of Jets")
plt.hist(ak.ravel(Mfrac_mu60),histtype='step',label='mu60',color='r',range=Range,bins=bins)
plt.hist(ak.ravel(Mfrac_mu200),linestyle='--',histtype='step',label='mu200',color='r',range=Range,bins=bins)
plt.yscale('log')
plt.grid(which='both')
plt.ylabel("Counts",loc='top')
plt.xlabel("Mfrac", loc='right')
plt.legend()
plt.show()
```
